Importing Dependencies

```python
import pandas as pd
```

Importing Dataset 

```python
df = pd.read_csv("D:\DA\Portfolio Projects\Supply Chain\smart_logistics_dataset.csv")

#Print the first 5 rows
print(df.head())
```

```python
#Check Datatypes
print(df.info())
```

Data Cleaning 

```python
#Convert into Datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
#Extract Time information
df['hour'] = df['Timestamp'].dt.hour
df['day'] = df['Timestamp'].dt.day_name()
df['month'] = df['Timestamp'].dt.month_name() 
```

```python
#Handle missing delay reasons
df['Logistics_Delay_Reason'] = df['Logistics_Delay_Reason'].fillna("No Delay")
```

Creating Features

```python
#Create Inventory Risk
df['Inventory_Risk'] = df['Inventory_Level'] < df['Demand_Forecast']
```

```python
#Create High Delay Flag
df['High_Risk_Delay'] = ((df['Traffic_Status'].isin(['Heavy','Detour']))&(df['Logistics_Delay']==1))
```

```python
#Create Efficiency Score
df['Efficiency_Score'] = df['Asset_Utilization']/(df['Waiting_Time']+1)
```
